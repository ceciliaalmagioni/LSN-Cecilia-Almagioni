# Esercitazione 8

Il principio variazionale in Meccanica Quantistica afferma che il valore di aspettazione dell'Hamiltoniana rispetto a un ragionevole stato quantistico di prova ha energia di stato fondamentale $E_0$ come limite inferiore:

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

Lo scopo di questa esercitazione è di ottimizzare lo stato fondamentale di una particella singola monodimensionale soggetta al seguente potenziale esterno:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

Il problema non è analiticamente risolubile dunque necessita di approcci numerici. Per ottenere una funzione d'onda approssimata il metodo Monte Carlo Variazionale utilizza la seguente funzione d'onda di prova (somma di due gaussiane) parametrzzata da una coppia di parametri $(\mu, \sigma)$:

$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Tramite l'ottimizzazione dei parametri $(\mu, \sigma)$ si cerca il minimo di $\langle\hat H\rangle_T$.


In questa esercitazione ho utilizzato come algoritmo di ottimizzazione il Simulated Annealing, che si ispira al processo di annealing termico dei materiali. Il suo obiettivo è trovare la configurazione che minimizza una funzione di costo, nota come energia, in un sistema fisico descritto da un'hamiltoniana. 
\
L'algoritmo del Simulated Annealing utilizza il concetto di temperatura virtuale.  Inizialmente, la temperatura è alta e l'algoritmo esplora ampiamente lo spazio delle soluzioni. Man mano che la temperatura diminuisce, l'algoritmo si concentra sempre più sul miglioramento delle soluzioni. Salti occasionali a livelli energetici più alti aiutano ad evitare minimi locali. Per ottenere ciò è necessario calibrare opportunamente i valori iniziali.



### Esercizio 8.1

In questo esercizio ho utilizzato lo stesso algoritmo di Metropolis per campionare la densità di probabilità $|\Psi_T^{\sigma,\mu}(x)|^2$ utilizzando una densità di probabilità uniforme $T(x_{new}|x_{old})$. Dopo aver fatto ciò, ho calcolato mediante data blocking il valore di aspettazione dell'Hamiltoniana:

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

dove ho utilizzato ($\hbar=m=1$).
\
Questo è servito poi nel prossimo step dell'esercitazione per poter calcolare il valore di aspettazione dell'energia data una coppia $(\mu, \sigma)$.

Il primo rappresenta il valore di aspettazione dell'Hamiltoniana, il secondo è un istogramma delle posizioni campionae, confrontato con l'espressione analitica della funzione d'onda di prova.

Da quanto detto prima sulle basse incertezze di $\langle\hat H\rangle$ mi aspettavo che non ci fosse compatibilità statistica con il valore atteso ed effettivamente è così.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 100

#---------------------------------------

plt.figure(figsize = (12,8))

histofill_blk = 1000                  
x = np.linspace(-3,3,100)      

mu = 0.8
sgm = 0.6

pos2 = np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es08//pos2.txt", delimiter=' ', unpack='true')
plt.hist(pos2, 300, density = "true", label = "Sampled positions", color='orange')
plt.ylabel('Frequenze', fontsize = 15)
plt.xlabel('Posizione', fontsize = 15)
plt.title("Campionamento posizioni", fontsize = 20, fontweight = "bold")
plt.grid(True)
plt.legend()

#-------------Soluzione matrici-----------------

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

# Plot a few things
scale = 0.3
plt.plot(x,(psi[0])**2, label='Soluzione matrici', color='r')
plt.xlim(-3,+3)
plt.legend()


plt.show()

```

La compatibilità tra l'istogramma e la funzione è buona ma non ottima.
