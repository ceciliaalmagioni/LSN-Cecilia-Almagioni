### Esercizio 1.1

In questa esercizione devo testare il generatore di numeri Pseudo-Random fornito a lezione. Per far ciò ho stimato il valore dell'integrale $\langle r \rangle = \int_{0}^{1}r dr$ calcolando la media dei numeri generati casualmente tra 0 e 1.

La media è stata calcolata con il *data blocking* nel seguente modo: ho generato M numeri casuali e li ho suddivisi in N blocchi da L elementi ciscuno. Ho stimato il valor medio in cciascun blocco e infine calcolato il valor medio totale mediando sulle singole medie di ciascun blocco L'incertezza è stata calcolata così: $$ \sigma^2_A := \langle (A-\langle A\rangle)^2\rangle = \langle A^2 \rangle -\langle A\rangle^2 $$ dove $$ \langle A^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 $$
infine l'incertezza statistica è data da
$$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N A^2_i - \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 \right]} $$




```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, y, err=np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es01/1.1/dati1.txt", unpack=True)

plt.errorbar(x, y, yerr=err, color='darkorange')
plt.xlabel('#blocks')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.title('Valore medio')
plt.show()
```

Nel grafico sopra riportato è illustrato il valor medio di *r* meno il valore atteso (1/2), così che mi aspetto che il grafico fluttui attorno allo zero. 

Nel mio codice ho utilizzato
- M=10e6
- N=10e2
- L=10e4
L'errore è stato stimato col metodo dei blocchi come illustrato sopra, sul valor medio di un blocco abbiamo considerato le stime di tutti i blocchi precedenti. In questo modo, l'incertezza statistica risulta diminuire con il progredire dei blocchi e il valor medio tende a stabilirsi su un valore compatibile con lo 0.

Analogamente, ho stimato il valore di  $\sigma^2 = \int_{0}^{1}(r-\frac{1}{2})^2 dr = \frac{1}{12}$

Per testare i risultati, ho effettuato un test del $\chi^2$. Ho suddiviso l'intervallo $[0, 1]$ in 100 intervalli e in ciascun intervallo sono stati effettuati 10000 lanci, ripetendo il test 100 volte.  

Successivamente abbiamo implementato un test del chi-quadro per testare l'effettiva randomicità del nostro generatore di numeri pseudo-casuali. Abbiamo diviso l'intervallo
in intervalli e abbiamo effettuato lanci. Abbiamo eseguito tale test per 100 volte.
Il $\chi^2$ calcolato per ciascun intervallo (rappresentato nel bin dell'istogramma) è dato dalla formula:  $$\chi^2 = \sum_{i=1}^{M} \frac{(n_i-n/M)^2}{n/M}$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data = np.loadtxt('/home/cecilia/Scrivania/LabSN/Esercitazioni/Es01/1.1/chi2.txt')
#plt.hist(data, bins=20)

plt.figure(figsize=(10, 5))
plt.suptitle('Test del $\chi^2$', fontweight='bold')

plt.subplot(1,2,1)
n, bins, patches = plt.hist(data, bins=20, density=True, alpha=0.75, color='darkgreen')
# Calcola la curva gaussiana
mu, sigma = np.mean(data), np.std(data)
x = np.linspace(bins[0], bins[-1], 100)
plt.xlim(60,140)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
# Disegna la curva gaussiana sopra l'istogramma
plt.plot(x, y, color='r', linewidth=2)
plt.xlabel('chi-square')
plt.ylabel('densità di probabilità')
plt.title('Chi-square test')
plt.grid()

plt.subplot(1,2,2)
x=np.arange(0,100)
plt.errorbar(x,data, color='darkgreen')
plt.hlines(100, 0, 100, color='r')
plt.xlabel('# iteration')
plt.ylabel('$\chi^2$')
plt.title('$\chi^2$')
plt.grid(True)
plt.show()
```

Nell'istogramma è riportato in rosso l'andamento della distribuzione di probabilità del $\chi^2$ per confronto. Il valor medio della distribuzione atteso è $(n_i-n/M)\simeq n/M$ cioè nel nostro caso $\chi^2 \simeq 100$.

### Esercizio 1.2

L'obiettivo di questo esercizio è di estendere il Generatore di Numeri Pseudo-Random per poter estrarre numeri casuali distribuiti secondo distribuzioni diverse da quella uniforme.

Verificherò inoltre la validità del Teorema del Limite Centrale sommando variabili generate casualmente secondo diverse distribuzioni, mostrando come le medie tendano a distribuirsi secondo una distribuzione normale centrata sul valore vero. Questo è vero solo se le distribuzioni con cui sono generate le varibili casuali hanno varianza finita.

Ho dunque aggiunto alla classe `Random`nuove possibilità di estrazione di numeri casuali:

- distribuzione esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$;
- distribuzione di Cauchy-Lorentz $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$

usando il metodo dell'inversione della Cumulative Distribution Function (CDF) ottenendo rispettivamente:
- $x = -\frac{1}{\lambda}\ln(1-y)$
- $x = \Gamma \tan(\pi(y-\frac{1}{2}))+\mu$

con $y$ estratto uniformemente in $[0,1]$.

Riporto in seguito gli istogrammi ottenuti riempiendo con $10^4$ realizzazioni di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$ con $x_i$ una variabile casuale ottenuta lanciando un dado standard, un dado esponenziale ($\lambda = 1$) e un dado lorentziano ($\mu =0$ e $\Gamma =1$).

### Esercizio 1.3

Nell'ultima parte di questa esercitazione ho calcolato il valore di $\pi$ tramite l'esperimento di Buffon.
Lo storico esperimento consiste nel lanciare un ago di lunghezza $L$ su un piano avente disegnate delle righe orizzontali equispaziate a distanza $d>L$. La probabilità che un ago intersechi una riga è $$P=\frac{2L}{\pi d}$$

Lanciando un ago $N_{throw}$ volte e colpendo $N_{hit}$ volte una riga, si può ricavare $\pi$ da 
$$\pi=\lim_{{N \to \infty}} \frac{2LN_{throw}}{N_{hit}d}$$

Ho generato $10^5$ lanci suddivisi in $100$ blocchi per il *data blocking* che serve a trovare valor medio di $\pi$ e la sua incertezza statistica analogamenta e quanto fatto nell'esercizio 1.1.
Per "lanciare" l'ago ho generato casualmente la posizione del baricentro (numero random tra $-d/2$ e $d/2$) e l'angolo di inclinazione rispetto alle linee parallele con metodo *accept-reject*. Quest'ultimo funziona nel seguente modo: viene generato un punto in un rettangolo $2$x$1$ del semipiano cartesiano superiore. Se le coordinate di questo punto appartengono al semicerchio di lato 1 ($x_p^2+y_p^2<1$), l'angolo è dato da 
$$\arccos(\frac{x_p}{\sqrt{x_p^2+y_p^2}})$$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

N = 100
x_block = np.arange(N)+1 
x, y, err=np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es01/1.3/pi.txt", unpack=True)

pi=math.pi
pivec = pi*np.ones(x.size)

plt.title('Calcolo di $\pi$')
plt.errorbar(x_block, y, yerr=err, color='yellowgreen')
plt.plot(x_block,pivec,'r--',label='True value')
plt.xlabel('#blocks')
plt.ylabel('$\pi$')
plt.grid(True)
#plt.title('Valore medio')
plt.show()

print('PI medio (atteso '+str(np.pi)+'): ')
print(y[-1])
print('\nErrore statistico : ')
print(err[-1])
```
