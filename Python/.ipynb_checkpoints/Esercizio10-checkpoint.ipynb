In questo esercizio ho parallelizzato la risoluzione dell'esercizio 9 del problema del commesso viaggiatore (TSP). Ho utilizzato la libreria MPI (Message Passing Interface) e il modello SPMD (Single Programme Multiple Data).

La struttura dell'algoritmo genetico è rimasta pressoché identica all'esercitazione 9 alla quale rimando per i dettagli sull'implementazione della ricerca genetica. Per parallelizzare il codice ho aaportato poche modifiche nel main:

- ho modificato il seme del generatore di numeri casuali differenziandolo in base al *rank*, così che i diversi processi non effettuano la stessa ricerca (altrimenti la aprallelizzazione sarebbe inutile);
- ho implementato la migrazione dei migliori elementi della popolazione di ciascun rank ogni $N_{migr}$ utilizzando `MPI_Send` e `MPI_Recv` per scambiarsi gli individui.

Per calibrare i parametri dell'algoritmo genetico e verificarne il corretto funzionamento ho testato il codice sul problema dell'esercizio 9 delle $34$ città disposte in un quadrato di raggio $2$.

Ho asciato i parametri invariati rispetto all'esercizio 9 così da poter avere un confronto:
- $N_{cities}=34$ $dimpop=1000$ $N_{gen}=500$
- $p_c=0.55$, $p_{m1}=0.3$, $p_{m2}0.1$ e $p_{m3}=0.1$ (ho tenuto $p_{m1}=0.3$ al di sopra del $10\%$ poichè la modifica apportata al cromosoma da `pair_permut` è molto piccola);
- $p=1.7$

In questo modo mi sono accorta che l'esercizio 9 non era ben risolto: probabilmente nel 9 sono incappata in un minimo locale e il maggior ricambio genetico dato dalla migrazione tra core è riuscito a proseguire la ricerca oltre a quel minimo. 
\
Questo fatto mostra evidentemente come la ricerca parallela (con migrazione) sia molto più efficiente di quella a singolo processo. Per trovare la soluzione corretta avrei dovuto aggiungere mutazioni e/o aumentare la probabilità delle stesse o del crossover, oppure avrei dovuto aumentare la dimensione della popolazione o il numero di generazioni.
