### Esercizio 4.1

Ci è stato fornito un codice già funzionante che sfrutta l'algoritmo di Verlet per simulare l'evoluzione dinamica di un sistema a $N=108$ particelle descritto da potenziale di Lennard-Jones e per calcolarne: energia totale per particella $E/N$, energia potenziale per particella $U/N$, energia cinetica per particella $K/N$ e temperatura $T$.

Il programma legge un file di input dal quale salva i valori per la densità $\rho$, il raggio di cut-off $r_{cut}$, il passo $\delta$, la temperatura iniziale $T$ diversi per ogni stato su cui si conduce la simulazione.

Per completare il codice ho dovuto implementare il calcolo della pressione $P$ in forma di medie di blocco, medie progressive e incertezze progressive.
Dal teorema del viriale, la pressione in un sistema di Lennard-Jones può essere stimata da:

$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

Durante tutta la simulazione ho utilizzato unità di misura ridotte: $U=\epsilon U^*$, $E=\epsilon E^*$, $K=\epsilon K^*$, $T=\frac{\epsilon}{k_B} T^*$, $P=\frac{\sigma^3}{k_B} P^*$. 
Per leggerezza di notazione d'ora in avanti ometterò gli asterischi dalle grandezze ridotte.
La pressione diventa così:

$$ P = \rho T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48 \left[ \left(\frac{1}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - \frac{1}{2} \left(\frac{1}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

Nel codice, il calcolo della media di ensemble è stato implementato in modo uniforme al calcolo della media di ensemble delle altre variabili.

**Equilibrazione**

La configurazione iniziale del sistema è quella di un cristallo, mentre le velocità sono generate casualmente con distribuzione di Maxwell-Boltzmann. Nonostante la configurazione iniziale sia di equilibrio, si osserva che la temperatura tende a deviare da quella impostatanel file di input. Questo fatto è dovuto ad una redistribuzione dell'energia iniziale ed è compatibile con il fatto che nell'ensemble microcanonico la temperatura non è fissata.

Si può risolvere questo fatto cambiando la temperatura iniziale nei parametri di input, scegliendone una che tenda a stabilizzarsi sul valore desiderato, lasciandola equilibrare per un numero adeguato di passi. Ho lasciato equilibrare il sistema solido e quello liquido per 20000 passi, mentre il sistema gas sembrava metterci di più a stabilizzare la temperatura, dunque ho calibrato per 50000 passi. 

Dopo di ciò si può iniziare la simulazione impostando il comando `Restart=1` così che il sistema riparta dalla configurazione equilibrata.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

npart = 108

N = 20000
N2 = 50000

phases = ["solid", "liquid", "gas"]
titles = ["Fase solida", "Fase liquida", "Fase gas"]
Teq = [0.8, 1.1, 1.2]

plt.figure(figsize=(21,15))
plt.suptitle("Equilibrazione della temperatura", fontsize = 17, fontweight = "bold")



plt.subplot(3,1,1)
x = np.arange(N)+1    # uso l'ordinamento da 1 a N
xS, aveTS, TS, errT=np.loadtxt("//home/cecilia/Scrivania/LabSN/Esercitazioni/Es04/MD_MC/MDNVE_solid/Eq_temp.dat", unpack=True)
plt.plot(x, aveTS, marker=None, label="Temperatura progressiva")
plt.title(titles[0], fontsize = 18)
plt.ylabel('T', fontsize = 15)
plt.xlabel('Step number', fontsize = 25)
plt.axhline(y=Teq[0], color ="r", label = "Temperatura target: "+str(Teq[0]))
plt.grid(True)
plt.legend()
plt.subplots_adjust(hspace=0.3)
    
    
plt.subplot(3,1,2)
xL, aveTL, TL, errTL=np.loadtxt("//home/cecilia/Scrivania/LabSN/Esercitazioni/Es04/MD_MC/MDNVE_liquid/Eq_temp.dat", unpack=True)
plt.plot(x, aveTL, marker=None, label="Temperatura progressiva")
plt.title(titles[1], fontsize = 18)
plt.ylabel('T', fontsize = 15)
plt.xlabel('Step number', fontsize = 15)
plt.axhline(y=Teq[1], color ="r", label = "Temperatura target: "+str(Teq[1]))
plt.grid(True)
plt.legend()
plt.subplots_adjust(hspace=0.3)
    
x2 = np.arange(N2)+1    
plt.subplot(3,1,3)
xG, aveTG, TG, errTG=np.loadtxt("//home/cecilia/Scrivania/LabSN/Esercitazioni/Es04/MD_MC/MDNVE_gas/Eq_temp.dat", unpack=True)
plt.title(titles[2], fontsize = 18)
plt.plot(x2, aveTG, marker=None, label="Temperatura progressiva")
plt.ylabel('T', fontsize = 15)
plt.xlabel('Step number', fontsize = 15)
plt.axhline(y=Teq[2], color ="r", label = "Temperatura target: "+str(Teq[2]))
plt.grid(True)
plt.legend()
plt.subplots_adjust(hspace=0.3)
    
    
plt.show()
```

**Osservazioni finali**

I grafici dell'energia totale per particella sembrano oscillare, in contrasto all'ipotesi della conservazione dell'enegia. Si nota però che queste oscillazioni avvengono su scala inferiore di un ordine di grandezza rispetto alle altre energie.

Il grafico dell'energia totale del gas risultava avere un comportamento anomalo (energia che rimaneva costate per un rilevante numero di blocchi). Il problema era dovuto al fatto che le oscillazioni erano così piccole da venire troncate. Questo è stato risolto richiedendo un troncamento a cifre decimali più elevate, nel mio caso 12, impostando il comando `setprecision(12)`.

Si osserva no le seguenti proporzionalità:

- l'andamento dell'energia cinetica proporzionale all'andamento della temperatura;
- il valore della pressione è proporzionale alla densità.
