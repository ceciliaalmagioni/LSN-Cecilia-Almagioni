### Esercizio 11.1 Fit lineare

In questo esercizio considero il caso di una funzione lineare $$ f(x) = 2x + 1 $$ eseguendo tentativi di fit su di essa con una rete neurale sequenziale.

Il codice fornito prevede l'implementazione di una rete a un solo neurone. Dal momento che un neurone riceve un input $ x $ e restituisce un output $ y= w\cdot x + b$, dove $w$ $b$ rappresentano un peso e un bias, che sono i parametri che il modello dovrà calibrare, un solo neurone è sifficiente per un fit lineare. 

In questo esercizio lo scopo è di studiare come cambia la qualità del fit e delle previsioni del modello, modificando il numero di epoche, $N_{epochs}$, il numero di dati di train $N_{train}$ e il rumore $\sigma$. Si è sempre mantenuto il numero di dati di train $N_{train}$ almeno dieci volte maggiore il numero di dati di validazione.



```python

import matplotlib.pyplot as plt
import numpy as np

# parametri da modificare
N_epochs = 20
N_train = 1000 # dati di train
N_valid = 50 # dati di test
Sigma = 0.2

# parametri target of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

np.random.seed(0)
x_train = np.random.uniform(-1, 1, N_train)        # input training 
x_valid = np.random.uniform(-1, 1, N_valid)        # input di test
x_valid.sort()
y_target = m * x_valid + b                         # target 

sigma = Sigma                                      # standard deviation (rumore)
y_train = np.random.normal(m * x_train + b, sigma) # array di misure da cui vogliamo indovinare i parametri di regressione
y_valid = np.random.normal(m * x_valid + b, sigma) # array di misure dalle quali vogliao testare abilità di predizione della NN

# validazione del plot validation e dataset target 
plt.figure(figsize=(8,5))
plt.plot(x_valid, y_target, label = "Target function\n f(x) = 2x+1")
plt.scatter(x_valid, y_valid, marker = ".", color='r', label='Validation data (for test)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Validation data and Target function", fontweight = "bold")
plt.grid()
plt.legend()
plt.show()

```

In questo grafico è mostrato un confronto tra la funzione target che vogliamo da fittare e i dati utilizati per il test di validazione. 

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))  #un neurone, con che forma 
                                        

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# Get a summary of our composed model:
# model.summary()

# TRAIN THE MODEL: do al neurone il set (x,y) di dati per il training
# L'optimizer trova il migliore weights minimizzando la funzione costo (MSE)

# Fit the model over 10 epochs of 32 batch size each.
# Report training progress against validation data
history = model.fit(    x=x_train, y=y_train, 
                        batch_size=32, epochs=N_epochs,
                        # shuffle input before each epoch (not mandatory)
                        shuffle=True,                        
                        validation_data=(x_valid, y_valid))
```

```python
# Return Weights and Biases (w <-> m=2)(b <-> q=1)

print("Guessed parameters")
w,b = model.get_weights() # hope they are similar to m=2 and q=1.
                          # Was the fit successful?
print("Slope   m = "+str(w))
print("Interc. q = "+str(b))
print()

# VALUTAZIONE DEL MODELLO

# valuta con TEST data
print("TEST DATA evaluation")
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
print('Test loss:     ', score[0]) #performance
print('Test accuracy: ', score[1])
print()

# valuta con la curva esatta EXACT
print("EXACT CURVE evaluation")
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
print('Test loss:     ', score[0])
print('Test accuracy: ', score[1])


plt.figure(figsize=(20,5))

# Plot training & validation loss values
plt.subplot(1,3,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss', fontweight = "bold")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
#plt.show()

# generate predictions
x_predicted = np.random.uniform(-1, 2, 100) 
y_predicted = model.predict(x_predicted)

# (re)plot validation and target dataset
plt.subplot(1,3,2)
plt.plot(x_valid, y_target, label = "Target function\n f(x) = 2x+1")
plt.scatter(x_valid, y_valid, marker = ".", color='r', label='Validation data (for test)')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

# plot
plt.subplot(1,3,3)
plt.scatter(x_predicted, y_predicted, marker = ".", color='r', label = "Model prediction")
plt.plot(x_valid, y_target, label = "Target function\n f(x) = 2x+1")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Model predictions and Target function", fontweight = "bold")
plt.grid()
plt.legend()
plt.show()
```

Grafico a sinistra: andamento delle loss (MeanSquareError) sui dati di calibrazione e su quelli di test, in funzione dell'epoca. Se la calibrazione è buona, entrabe convergono ad un valore basso in modo approssimabile ad una curva esponenziale. 

Grafico centrale: come il grafico sopra.

Grafico a destra: predizioni del modello su un dominio esteso rispetto a quello su cui è stata eseguita la calibrazione. La calibrazione è stata efficace perche le previsioni sono in accordo con la legge lineare anche al di fuori del dominio di train.

Sperimentando con i parametri iniziali, si osserva che più i dati di train sono rumorosi, maggiore è il numero di epoche per avere una buona predizione.

Aumentando il numero di dati di training, la rete apprende più velocemente. Fissato il rumore, infatti, si osserva che per $N_{train}=1000$, si raggiunge un valore basso delle funzioni costo (train e test) in 10 epoche, mentre per $N_{train}=500$ sono necessarie almeno 20 epoche. 
In un problema come questo, risulta più conveniente lavorare con più dati di training (un migliaio) e poche epoche (qualche decina).

### Esercizio 11.2

In questo esercizio ho ceracto di estendere il modello dell'esercizio precedente in modo da poter fittare le seguente funzione polinomiale del terzo ordine: 

$$
f(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.

La funzione è più complessa della precedente, dunque un singolo neurone non è più sufficiente. Ho tentato dunque di variare il numero di layer e di neuroni per layer. Ho inoltre modificato $N_{epochs}$. 




```python
import matplotlib.pyplot as plt
import numpy as np


# parametri
N_epochs = 20
N_train = 5000 # <<< train
N_valid = 50 # <<< test
Sigma = 0.1

# parametri target 

# f(x) = 3x^3 - 2x^2 - 3x + 4 
# f(x) = ax^3 + bx^2 + cx + d
a = 3 
b = -2
c = -3
d = 4
def f(x):
    return a*x**3 + b*x**2 + c*x + d

#generazione dati input 
np.random.seed(0)
x_train = np.random.uniform(-1, 1, N_train)       
x_valid = np.random.uniform(-1, 1, N_valid)       
x_valid.sort() 
y_target = f(x_valid)                             

sigma = Sigma                                      # rumore
y_train = np.random.normal(f(x_train), sigma)      
y_valid = np.random.normal(f(x_valid), sigma)      

# plot dataset validazione e target
plt.figure(figsize=(8,5))
plt.scatter(x_train, y_train, marker = "x", color='orange', label='Training data',linewidth=0.5, alpha=0.4)
plt.plot(x_valid, y_target, label = "Target function\n $f(x) = 3x^3 - 2x^2 - 3x + 4$")
plt.scatter(x_valid, y_valid, marker = ".", color='g', label='Validation data (for test)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Validation data and Target function", fontweight = "bold")
plt.grid()
plt.legend()
plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
# activation: relu, elu, sigmoid, softmax, softplus, softsign, tanh, exponential
act = 'relu'
model.add(Dense(40, input_shape=(1,), activation = act))
model.add(Dense(10, activation = act))
model.add(Dense(5, activation = act))
model.add(Dense(1, activation = act))

#Vengono aggiunti strati Dense al modello con diversi numeri di unità e la funzione di attivazione specificata. 
#L'ultimo strato ha un'unica unità perché si tratta di un problema di regressione

# compile the model choosing optimizer, loss and metrics objects
# optimizers: sgd, rmsprop, adam, adadelta, adagrad, adamax, nadam, ftrl
opt = 'rmsprop'
# loss functios: mse, mae, msle, logcosh, kld
lf = 'mse'
model.compile(optimizer=opt, loss=lf, metrics=['mse'])

# Get a summary of our composed model
model.summary()


# Fit the model over 10 epochs of 32 batch size each.
# Report training progress against validation data
history = model.fit(    x=x_train, y=y_train, 
                        batch_size=32, epochs=N_epochs,
                        # shuffle input before each epoch (not mandatory)
                        shuffle=True,                        
                        validation_data=(x_valid, y_valid))

```

```python

# Return Weights and Biases
#print("Guessed parameters")
#w = model.get_weights()
#print(w)

# EVALUATE MODEL

# evaluate with TEST data
print("TEST DATA evaluation")
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
print('Test loss:     ', score[0])
print('Test accuracy: ', score[1])
print()

# evaluate with the EXACT curve
print("EXACT CURVE evaluation")
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
print('Test loss:     ', score[0])
print('Test accuracy: ', score[1])

# Training history

plt.figure(figsize=(20,5))

# Plot training & validation loss values
plt.subplot(1,3,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss', fontweight = "bold")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
#plt.show()


# COMPARE MODEL predictions with TARGET FUNCTION

# generate predictions
x_predicted = np.random.uniform(-1.5, 1.5, 100) 
y_predicted = model.predict(x_predicted) 

# (re)plot validation and target dataset
plt.subplot(1,3,2)
plt.scatter(x_train, y_train, marker = "x", color='orange', label='Training data',linewidth=0.5, alpha=0.4)
plt.plot(x_valid, y_target, label = "Target function\n $f(x) = 3x^3 - 2x^2 - 3x + 4$")
plt.scatter(x_valid, y_valid, marker = ".", color='g', label='Validation data (for test)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Validation data and Target function\n(from before)", fontweight = "bold")
plt.grid()
plt.legend()

# plot
plt.subplot(1,3,3)
plt.scatter(x_train, y_train, marker = "x", color='orange', label='Training data',linewidth=0.5, alpha=0.4)
plt.scatter(x_predicted, y_predicted, marker = ".", color='g', label = "Model prediction")
x_predicted.sort()
y_target_ext = f(x_predicted) # target polinomial function
plt.plot(x_predicted, y_target_ext, label = "Target function\n $f(x) = 3x^3 - 2x^2 - 3x + 4$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Model predictions and Target function", fontweight = "bold")
plt.grid()
plt.legend()
plt.show()
```

Grafico a sinistra: andamento delle loss (MeanSquareError) sui dati di calibrazione e su quelli di test, in funzione dell'epoca. Se la calibrazione è buona, entrabe convergono ad un valore basso in modo approssimabile ad una curva esponenziale. 

Grafico al centro: stesso grafico di sopra, per memo. 

Grafico a destra: predizioni del modello su un dominio esteso rispetto a quello su cui è stata eseguita la calibrazione. In questo caso risulta molto fifficile ottenere una buona predizione anche al di fuori del dominio di train.

**Osservazioni**

L'obiettivo è stato quello di avere una accuracy dello 0.01 circa.

Si osserva che dopo circa 6 epoche la funzione costo non migliora significativamente.

All'interno del dominio  $[-1, 1]$ la predizione è molto efficace, mentre al di fuori del dominio non si riesce ad ottenere una predizione compatibile con la funzione target. 

### Esercizio 11.3


In questo esercizio proviamo ad estendere il modello per fittare una funzione trigonometrica bidimensionale
 $$f(x,y) = \sin(x^2+y^2)$$
nel dominio $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.


Dato che la funzione ha  in ingrasso due variabili, abbiamo modificato ???


```python
import matplotlib.pyplot as plt
import numpy as np

# parametri
N_epochs = 50
N_train = 10000 # train
N_valid = 1000 # test
Sigma = 0.1


np.random.seed(0)
lim = 3./2.
x_train = np.random.uniform(-lim, lim, N_train)    # training inputs: funzione bidimensionale
x_valid = np.random.uniform(-lim, lim, N_valid)    # 

y_train = np.random.uniform(-lim, lim, N_train)    # training inputs
y_valid = np.random.uniform(-lim, lim, N_valid)    # ...

XY_train = np.column_stack((x_train,y_train))      # MATRICE PER DATI BIDIMENSIONALI
XY_valid = np.column_stack((x_valid,y_valid))      

def f(x, y):
    return np.sin((x ** 2 + y ** 2))

z_target = f(x_valid, y_valid)                    

sigma = Sigma                                      
z_train = np.random.normal(f(x_train,y_train), sigma) 
z_valid = np.random.normal(f(x_valid,y_valid), sigma)
                                                   

# plot validation and target dataset
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (7,7))
ax = Axes3D(fig)
fig.add_axes(ax)

x_surf = np.arange(-lim,lim, 0.05)
y_surf = np.arange(-lim,lim, 0.05)
X_surf, Y_surf = np.meshgrid(x_surf,y_surf)
ax.scatter(x_train, y_train, z_train, marker = "x", color = "orange",
               label='Training data', linewidth=0.3, alpha = 0.4)
ax.plot_wireframe(  X_surf, Y_surf, f(X_surf,Y_surf), 
                    rstride=2, cstride=2,color = 'r', 
                    label='Target function\n $\sin(x^2+y^2)$',
                    alpha = 0.5)
ax.scatter(x_valid, y_valid, z_valid, marker = "*", color = "g", label='Validation data (for test)')
ax.set_xlabel("x", fontsize = 18)
ax.set_ylabel("y", fontsize = 18)
ax.set_zlabel("z", fontsize = 18)
plt.title("Validation data and Target function", fontsize = 18, fontweight = "bold")
ax.legend()
ax.view_init(10, 30)

plt.show()
```

```python
# INITIALIZE Neural Network (Sequential) model:

# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
# activation: reluNO, eluSI, sigmoid-terribile, softmax-??, softplus-ni, softsign, tanhSI, exponential-non funz
act = 'tanh'
model.add(Dense(45, input_shape=(2,), activation = act)) 
model.add(Dense(40, activation = act))
model.add(Dense(25, activation = act))
model.add(Dense(10, activation = act))
model.add(Dense(1, activation = act))

# compile the model choosing optimizer, loss and metrics objects
# optimizers: sgd, rmsprop, adam, adadelta, adagrad, adamax, nadam-si con tanh, ftrl
opt = 'nadam'
# loss functios: mse, mae, msle, logcosh, kld
lf = 'mse'
model.compile(optimizer=opt, loss=lf, metrics=['mse'])

# Get a summary of our composed model
model.summary()

# TRAIN THE MODEL 

# Fit the model over 10 epochs of 32 batch size each.
# Report training progress against validation data
history = model.fit(    x=XY_train, y=z_train, 
                        batch_size=32, epochs=N_epochs,
                        # shuffle input before each epoch (not mandatory)
                        shuffle=True,                        
                        validation_data=(XY_valid, z_valid))

```

```python
# Return Weights and Biases

# print("Guessed parameters")
# w = model.get_weights()
# print(w)


# EVALUATE MODEL

# evaluate with TEST data
print("TEST DATA evaluation")
score = model.evaluate(XY_valid, z_valid, batch_size=32, verbose=0)
print('Test loss:     ', score[0])
print('Test accuracy: ', score[1])
print()

# evaluate with the EXACT curve
print("EXACT CURVE evaluation")
score = model.evaluate(XY_valid, z_target, batch_size=32, verbose=0)
print('Test loss:     ', score[0])
print('Test accuracy: ', score[1])

# Training history

plt.figure(figsize=(7,5))

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss', fontweight = "bold")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
plt.show()

# COMPARE MODEL predictions with TARGET FUNCTION

# generate predictions
x_predicted = np.random.uniform(-lim, lim, 2000) 
y_predicted = np.random.uniform(-lim, lim, 2000) 
XY_predicted = np.column_stack((x_predicted,y_predicted))
z_predicted = model.predict(XY_predicted)

# plot
fig = plt.figure(figsize = (7,7))
ax = Axes3D(fig)
fig.add_axes(ax)
ax.scatter(x_train, y_train, z_train, marker = "x", color = "orange",
               label='Training data', linewidth=0.3, alpha = 0.4)
ax.plot_wireframe(  X_surf, Y_surf, f(X_surf,Y_surf), 
                    rstride=2, cstride=2, color = 'r', 
                    label='Target function\n $\sin(x^2+y^2)$',
                    alpha = 0.2)

ax.scatter(x_predicted, y_predicted, z_predicted, marker = "*", color = "g", label='Validation data (for test)')
ax.set_xlabel("x", fontsize = 18)
ax.set_ylabel("y", fontsize = 18)
ax.set_zlabel("z", fontsize = 18)
plt.title("Validation data and Target function (from before)", fontsize = 18, fontweight = "bold")
ax.legend()
ax.view_init(10, 30)


plt.show()
```
