### Esercizio 2.1

In questo esercizio ho calcolato l'integrale $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ con due metodi Monte Carlo differenti:
- campionando con una distribuzione uniforme in $[0, 1]$
- con importance sampling, cioè campionando con una distribuzione non uniforme in $[0,1]$ $p(x)$ scelta accuratamente.

La scelta della $p(x)$ per il metodo dell'importance sampling è importante: con un'opportuna $p(x)$ si campionano maggiormente le regioni in cui la funzione da integrare è significativamente diversa da zero. 
Nel dominio di integrazione dunque $p(x)$ deve assomigliare all'integranda, essere positiva, finita, normalizzata e la sua CDF dev'essere invertibile. 
Con uesto metodo è possibile ridurre la varianza rispetto al campionamento uniforme.   
La più semplice funzione rispettante le sopracitate caratteristiche è risultata essere $$p(x)=2(1-x)$$la cui CDF è $y=2x-x^2$, con inversa $$x=1-\sqrt(1-y)$$ dove y è campionato uniformemente in $[0,1]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure(figsize=(10,7))
x = np.linspace(0., 1., 100)
f = np.cos(math.pi/2. * x)
P = 2*(1 - x) 
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.plot(x, math.pi/2. * f, label='Integranda')
plt.plot(x, P, label ='p(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Esercizio 2.2

In questo secondo esercizio ho simulato due Random Walk tridimensionali con passo costante in modulo $a=1$.

- Random Walk discreto: i passi avvengono possono avvenire solo lungo uno dei tre assi cartesiani, sia in avanti che indietro;

- Random Walk continuo: i passi possono avvenire in una direzione casuale nello spazio. 

Per entrambi eseguo $10^4$ Random Walk da 100 passi, calcolando ad ogni passo $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$, dove $\vec{r}_N$ rappresenza la distanza dall'origine del passo N-esimo. 
Riporto dunque nei grafici sottostanti $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione del passo N-esimo ($N\in[0, 100]$), confrontando l'andamento dei punti del plot con la funzione $f(N)=k\sqrt(N)$ per verificare la diffusività del suo andamento.

**Random Walk discreto**

Per decidere la direzione dell'*i-esimo* passo ho estratto un intero in $[0,6)$: ogni intero è associato ad uno spostamento avanti oppure indietro sulle tre direzioni cartesiane (dunque $3$x$2=6$ possibilità) e il passo è univocamente determinato.

**Random Walk continuo**

Per decidere la direzione dell'*i-esimo* passo ho estratto due variabili casuali $\theta$ in $[0, \pi]$ e $\phi$ in $[0, 2\pi]$ e da questi ho calcolato le coordinate cartesiane $x, y, z$ del passo. 

Per ogni passo di ciascun RW ho salvato la distanza al quadrato dall'origine $r^2_N=x^2+y^2+z^2$, così che ho potuto calcolare la media tra i $10^4$ valori di $r^2_N$ su ciascun blocco, prenderne la radice quadrata e mediare sui blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def f (x, k):
    return k * np.sqrt(x)


xD, yD, errD=np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es02/2.2/RWdisc.txt", unpack=True)

xC, yC, errC=np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es02/2.2/RWcont.txt", unpack=True)

#discrete fit
p_opt_disc, p_cov_disc = curve_fit(f, xD, yD)
y_fit = f(x,p_opt_disc[0])

print( ' --- Random Walk over a discrete lattice --- ')
print('fit with k*sqrt(x): k = ',p_opt_disc)
print('confidence over k : +- ',np.sqrt(p_cov_disc))

#continuous fit
p_opt_cont, p_cov_cont = curve_fit(f, xC, yC)

print( ' --- Random Walk in a continuum space --- ')
print('fit with k*sqrt(x): k = '+str(p_opt_cont))
print('confidence over k : +- '+str(np.sqrt(p_cov_cont)))

#graphing
plt.figure(figsize=(12,14))

plt.subplot(211)
plt.errorbar(xD,yD,errD,fmt='.',label='MonteCarlo data', markersize=1)
plt.grid(True)
plt.xlabel('steps',fontsize=13)
plt.ylabel(r'$\sqrt{\langle|\vec{r}_{N}|^2\rangle}$',fontsize=13)
plt.title("Random Walk discreto",fontsize=16)
y_fitD = f(xD,p_opt_disc)
plt.plot(xD,y_fitD,'r',linewidth=2,label='$f(x)=k\sqrt{x}$', alpha=0.4) # plotting fitted function
plt.legend()

plt.subplot(212)
plt.errorbar(xC,yC,errC,fmt='.',label='MonteCarlo data', markersize=1)
plt.grid(True)
plt.xlabel('steps',fontsize=13)
plt.ylabel(r'$\sqrt{\langle|\vec{r}_{N}|^2\rangle}$',fontsize=13)
plt.title("Random Walk continuo",fontsize=16)
y_fitC = f(xC,p_opt_cont)
plt.plot(xC,y_fitC,'r',linewidth=2,label='$f(x)=k\sqrt{x}$', alpha=0.4) # plotting fitted function
plt.legend()


plt.show()


print("Mean value of the distance from the origin for discrete RW: ", yD[99],  'pm'  , errD[99])
print("Mean value of the distance from the origin for continuum RW: ", yC[99],  'pm'  , errC[99])
```

Entrambi i grafici sono compatibili con un comportamento diffusivo.
