# Esercitazione 9

In questo esercizio ho risolto il problema del commesso viggiatore (TSP) con un algoritmo genetico. Il commesso viaggiatore deve viaggiare in $N$ città dalle posizioni $\vec{c}_i$. Deve visitarle tutte partendo da una certa città e ritornando nella stessa minimizzando la lunghezza totale del viaggio.
\
Come funzione di costo ho considerato la distanza al quadrato:

$$
L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2+|\vec{x}_N-\vec{x}_{1}|^2
$$
dove il secondo termine è dovuto alla richiesta di tornare alla città di prtenza.
\
Assegnata ad ogni città un'etichetta da $0$ a $N-1$, un particolare percorso è rappresentato da un vettore di interi i cui elementi rappresentano le città nell'ordine in cui vengono visitate.

**Algoritmo genetico e implementazione**

Gli algoritmi genetici sono basati su concetti di selezione naturale, crossover e mutazione. Mediante la creazione di una popolazione iniziale di soluzioni candidate rappresentate come cromosomi, l'algoritmo genetico evolve questa popolazione nel corso di diverse generazioni per trovare soluzioni sempre migliori. La selezione favorisce i cromosomi più adatti, il crossover combina le caratteristiche dei genitori per generare nuove soluzioni e la mutazione introduce variazioni casuali per evitare la convergenza prematura.


Sinteticamente, l'algoritmo che ho implementato parte da un insieme possibile di percorsi, una matrice *dim_pop* x $N_{cities}$ chiamata *generazione 0*, creato nella funzione `Generate_pop`. Il primo elemento di *gen0* è il vettore di interi ordinato da $0$ a $N_{cities}-1$. I successivi elementi di $gen0$ sono ottenuti tramite permutazioni del primo elemento. 
\
Ho implementato una funzione `fitness` per il calcolo della lunghezza $L^{(2)}$ di ciascun percorso e, con un algoritmo *MergeSort* ho implementato una funzione `order` che ordina gli elementi di una generazione in ordine di $L^{(2)}$ crescente.
\
Ho fatto evolvere la *gen0* con la funzione `new_gen` che è il cuore dell'algoritmo genetico:

- tramite l'operatore di selezione `selector` ho scelto tra gli elementi della vecchia generazione un padre e una madre da far riprodurre. L'operatore di selezione favorisce gli elementi con un *fitness* migliore. L'indice $j$ del genitore è scelto così: $j=int(dimpop\cdot()1-r^p)+1$ dove $r\in[0, 1)$ e $p$ un opportuno esponente;
- ho fatto riprodurre i genitori con la funzione `crossover` con una certa probabilità $p_c$;
- ho applicato ai figli con probabilità $p_{m1}$, $p_{m2}$ e $p_{m3}$ mutazioni genetiche (`pair_permut`, `m_permut`, `invert`);

Ho ripetuto queste operazioni fino ad ottenere una nuova popolazione, l'ho ordinata in base al fitness e che a sua volta ho fatto evolvere per un totale di $N_{gen}$ generazioni.

Parametri utilizzati:
- $N_{cities}=34$ $dimpop=1000$ $N_{gen}=500$
- $p_c=0.55$, $p_{m1}=0.3$, $p_{m2}0.1$ e $p_{m3}=0.1$ (ho tenuto $p_{m1}=0.3$ al di sopra del $10\%$ poichè la modifica apportata al cromosoma da `pair_permut` è molto piccola);
- $p=1.7$.





### Esercizio 9.1

In questo esercizio la posizione delle città è disposta su una circonferenza di raggio $1$ (funzione `Pos_circle`). Per questa configurazione è facile verificare che la ricerca vada a buon fine in quanto il miglior percorso percorre la circonferenza. 
\
(Attenzione: la lunghezza attesa non è confrontabile con la lunghezza della circonferenza $2\pi$ in quanto le lunghezze sono calcolate con norma $L^{(2)}$).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


##L miglior percorso
x, f = np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es09/circle/bestL.txt", unpack='true',  delimiter=' ')
x50, f50 = np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es09/circle/besthalfL_circle.txt", unpack='true',  delimiter=' ')

plt.figure(figsize=(15, 5))
plt.plot(x,f, label = "Migliore")
plt.plot(x50,f50, label = "Migliore metà")
plt.xlabel('generazione', fontsize=18)
plt.ylabel('$L^{(2)}$', fontsize=18)
plt.title("Lunghezza dei percorsi", fontweight='bold', fontsize=22)
plt.axhline(y=f[499], color = "k", linestyle = "--", label = "Final best: "+str(f[499]))
plt.grid(True)
plt.legend()
plt.show()
print('Percorso migliore L2 = ', f[-1])



##percorso iniziale
plt.figure(figsize=(10, 5))

plt.subplot(121)
x, f = np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es09/circle/initial_path.txt", unpack='true',  delimiter=' ')
plt.plot(x,f, 'g', linewidth='0.5', label='Percorso')
plt.plot(x,f, 'g^', label='Città')
plt.scatter(x[0], f[0], c='r', marker='o', s=100, label='Punto di partenza')
plt.title('Percorso di partenza', fontweight='bold', fontsize=20)
plt.legend()
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)


##miglior percorso
plt.subplot(122)
x, f = np.loadtxt("/home/cecilia/Scrivania/LabSN/Esercitazioni/Es09/circle/final_path.txt", unpack='true',  delimiter=' ')
plt.plot(x,f, 'g', linewidth='0.5', label='Percorso')
plt.plot(x,f, 'g^', label='Città')
plt.scatter(x[0], f[0], c='r', marker='o', s=100, label='Punto di partenza')
plt.title('Miglior percorso', fontweight='bold', fontsize=20)
plt.legend()
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)
plt.show()
```

Conoscendo dunque la soluzione, ho potuto sceglere un numero adeguato di generazioni ed una dimensione adeguata della popolazione per affrontare il problema successivo, nel quale era più difficile riconscere la migliore soluzione. 
\
Un'altra possibile scelta sarebbe stata quella di aumentare il coeffiiente $p$ dell'operatore di selezione e diminuire il numero di generazioni, in quanto la convergenza al miglior percorso era molto più rapida. Infatti, ho osservato che con $p\sim 3$ lo scarto tra il miglior cromosoma di una generazione e la migliore metà degli elementi della stessa generazione era estremamente ridotto, che significava avere una scelta molto efficiente dei genitori più adatti alla riproduzione. Tuttavia, ho scelto di mantenere la ricerca più lenta e meno efficiente per essere sicura di esplorare sufficientemente tutte le possibilità (ricerca più ergodica).
