**Option pricing**

In questa esercitazione, ho condotto un'analisi numerica da confrontare con la soluzione analitica del problema dell'Option Pricing, secondo la teoria di Black-Scholes, utilizzata per determinare correttamente i prezzi delle opzioni europee di tipo Call e Put. Tale teoria considera che il prezzo di un asset nel tempo segua un modello stocastico simile a un Moto Browniano Geometrico (GBM), con un tasso di interesse risk-free costante $r$ (corrispondente al drift browniano) e una determinata volatilità $\sigma$.

Nel campo della finanza, un'opzione rappresenta un contratto tra due o più parti, in cui una parte assume un impegno mentre l'altra parte ottiene il diritto (e non l'obbligo) di acquistare o vendere il titolo. Un tipo comune di opzione è l'opzione europea, in cui il venditore (A) dell'opzione garantisce all'acquirente (B) il diritto di acquistare o vendere un determinato asset a un prezzo fissato in un momento specifico (la scadenza del contratto). È importante notare che solo il venditore A è obbligato dal contratto, mentre l'acquirente B può decidere di esercitare il diritto di acquistare o vendere l'asset solo se può trarne un profitto.

Il profitto è possibile solo se al momento della scadenza $T$ il prezzo di mercato dell'asset $S(T)$ è superiore al prezzo fissato $K$. In tal caso, l'acquirente B esercita un'opzione di tipo Call (acquistando l'asset), ottenendo un profitto pari a $S - K$. Al contrario, se il prezzo di mercato $S(T)$ è inferiore a $K$, l'acquirente B desidererà esercitare un'opzione di tipo Put (vendendo l'asset), ottenendo un profitto pari a $K - S$.


La soluzione analitica di Black-Scholes per l'andamento di $S(T)$ con valore iniziale $S(0)$ è data da:

$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    dove $W(t)\sim\mathcal{N}(0,t)$

Inoltre, dato che gli incrementi di $W$ sono indipendenti e distribuiti normalmente, c'è una semplice formula ricorsiva per stimare il valore di $S(t)$:

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$ e $t_i$ istanti di tempo successivi in $[0, t]$
    
    
Dato $S(0)$, è possibile dunque ottenere una stima di $S(t)$ in due modi differenti. Il primo è utilizzando direttamente la prima formula. Il secondo discretizzando il moto browniano geometrico, suddividendo l'intervallo $[0, t]$ in N sottointervalli usando così la seconda formula data.

In entambi casi ho calcolato i valori della Call e della Put con le seguenti formule:

- $C[S(0), 0]=\exp(-rT)\max(0, S-K)$;
- $P[S(0), 0]=\exp(-rT)\max(0, K-S)$

Il fatto che valutiamo il massimo indica proprio la possibilità di B di esercitare o meno l'opzione. Il prezzo finale va però normalizzato a causa del tasso di interesse applicato dalla banca.

Ecco i parametri utilizzati per la simulazione:

- prezzo dell'asset a $t=0$: $S(0)=100$
- scadenza: $T=1$
- prezzo da contratto: $K=100$
- tasso di interesse risk-free: $r=0.1$
- volatilità: $\sigma=0.25$

Ho trattato il problema generando $10^4$ Random Walk e ho suddiviso i dati in 100 blocchi per trovare i valori finali con data blocking.  


```python
#trovo il valore analitico di call e put

import scipy
from scipy.special import erf
import numpy
from numpy import exp
from numpy.lib.scimath import log, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == SOLUZIONE ANALITICA BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```
